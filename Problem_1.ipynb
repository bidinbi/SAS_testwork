```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data_prep.tab', delimiter='\t')
```

Импортим нужные библиотеки и считываем данные. Смотрим на них.

```python
data
```

Создаем массив с названиями столбцов, для которых нужно создать 2 дополнительных столбца.

```python
ar = data.columns[2:]
ar
```

Теперь для каждого элемента этого массива создаем два новых столбца по описанию задачи. У первого из дополнительных столбцов будет пропущенным первое значение, а остальные будут сдвигаться вниз из столбца-источника. Для второго аналогично.

```python
for column in ar:
    data[column + '_1'] = [np.nan] + list(data[column])[:-1]
    data[column + '_2'] = [np.nan, np.nan] + list(data[column])[:-2]
```

Сделаем так, чтобы столбы-источники и дополнительные столбцы шли друг за другом. Для этого создадим новый массив столбцов и заполним его нужным образом. Далее, просто меняем порядок столбцов у датафрейма.

```python
ar1 = []
for i in ar:
    ar1.append(i)
    ar1.append(i + '_1')
    ar1.append(i + '_2')
```

```python
data = data[['SEGM_DATE', 'CLIENT_ID'] + ar1]
data
```
